# Face detector

```python
import face_recognition
from PIL import Image, ImageDraw
import numpy as np # Import the numpy library

# Load the image and convert to RGB
image_path = "face2.jpg"
pil_image = Image.open(image_path).convert("RGB")

# Convert the PIL Image object to a NumPy array
# The face_recognition library requires a NumPy array as input
numpy_image = np.array(pil_image)

# Find all face locations in the image
face_locations = face_recognition.face_locations(numpy_image)

# Create a drawing object
draw = ImageDraw.Draw(pil_image)

# Draw rectangles around the detected faces
for top, right, bottom, left in face_locations:
    draw.rectangle([left, top, right, bottom], outline="green", width=3)

# Display or save the image
pil_image.show()
pil_image.save("faces_detected.jpg")

print(f"Found {len(face_locations)} face(s) in the image.")
```

# Plain Comparison

```python
import face_recognition
from PIL import Image
import numpy as np
from scipy.spatial.distance import euclidean

# --- IMPORTANT: Make sure to install the necessary libraries ---
# You may need to run:
# pip install face-recognition numpy Pillow scipy scikit-image

# --- Part 1: Load and process the two images ---
# Replace "image1.jpg" and "image2.jpg" with your actual image file paths
image_path_1 = "face3.jpg"
image_path_2 = "face4.jpg"

try:
    # Load the images and convert to NumPy arrays
    pil_image_1 = Image.open(image_path_1).convert("RGB")
    numpy_image_1 = np.array(pil_image_1)
    
    pil_image_2 = Image.open(image_path_2).convert("RGB")
    numpy_image_2 = np.array(pil_image_2)

    print("Images loaded successfully.")
except FileNotFoundError:
    print("Error: One or both image files were not found. Please check your paths.")
    exit()

# --- Part 2: Find face encodings for both images ---
# Face encodings are the 128-dimensional measurements of a face
encodings_1 = face_recognition.face_encodings(numpy_image_1)
encodings_2 = face_recognition.face_encodings(numpy_image_2)

# Check if a face was found in each image
if not encodings_1:
    print("No face found in the first image.")
    exit()
if not encodings_2:
    print("No face found in the second image.")
    exit()

# We will use the first face found in each image for comparison
face_encoding_1 = encodings_1[0]
face_encoding_2 = encodings_2[0]

# --- Part 3: Compare the faces using different metrics ---

# Method 1: Euclidean Distance
# This measures the "distance" between the two 128-dimensional vectors.
# A smaller distance indicates higher similarity.
euclidean_distance = euclidean(face_encoding_1, face_encoding_2)

# Method 2: Mean Squared Error (MSE)
# This calculates the average squared difference between the elements of the vectors.
# A smaller value indicates higher similarity.
# We'll use NumPy's vectorized operations for efficiency.
mse_score = np.mean((face_encoding_1 - face_encoding_2)**2)

# --- Part 4: Display the results ---
print("\n--- Face Comparison Results ---")
print(f"Comparison between '{image_path_1}' and '{image_path_2}':")
print(f"Euclidean Distance: {euclidean_distance:.4f}")
print(f"Mean Squared Error (MSE): {mse_score:.6f}")
print("\nNote: Lower values for both scores indicate higher similarity between the faces.")

# Optional: You can add the original face detection drawing code back here
# face_locations_1 = face_recognition.face_locations(numpy_image_1)
# pil_image_1 = Image.fromarray(numpy_image_1)
# draw1 = ImageDraw.Draw(pil_image_1)
# for top, right, bottom, left in face_locations_1:
#     draw1.rectangle([left, top, right, bottom], outline="green", width=3)
# pil_image_1.show()
```

# Comparison between multiple images

```python
import face_recognition
from PIL import Image
import numpy as np
from scipy.spatial.distance import euclidean

# --- IMPORTANT: Make sure to install the necessary libraries ---
# You may need to run:
# pip install face-recognition numpy Pillow scipy scikit-image

# --- Part 1: Define the target and candidate images ---
# The target image is the face you want to compare against all others.
target_image_path = "beast.png"

# The candidate images are the ones you want to check for a similar face.
candidate_image_paths = [
    "person1.jpg",
    "person2.jpg",
    "face3.jpg",
    "face4.jpg",
    "face1.jpg",
    "face2.jpg"
]

print(f"Comparing the face in '{target_image_path}' with faces in the following images:")
for path in candidate_image_paths:
    print(f"- {path}")

# --- Part 2: Find the face encoding for the target image ---
try:
    pil_target_image = Image.open(target_image_path).convert("RGB")
    numpy_target_image = np.array(pil_target_image)
    target_encodings = face_recognition.face_encodings(numpy_target_image)

    if not target_encodings:
        print(f"\nError: No face found in the target image '{target_image_path}'. Exiting.")
        exit()

    target_face_encoding = target_encodings[0]
    print("\nTarget face encoding generated successfully.")
except FileNotFoundError:
    print(f"\nError: Target image file '{target_image_path}' not found. Please check your path.")
    exit()

# --- Part 3: Loop through candidate images and compare faces ---
print("\n--- Comparison Results ---")
for candidate_path in candidate_image_paths:
    print(f"\nComparing with '{candidate_path}'...")
    try:
        pil_candidate_image = Image.open(candidate_path).convert("RGB")
        numpy_candidate_image = np.array(pil_candidate_image)
        candidate_encodings = face_recognition.face_encodings(numpy_candidate_image)
    except FileNotFoundError:
        print(f"Error: Candidate image file '{candidate_path}' not found. Skipping.")
        continue
    
    if not candidate_encodings:
        print(f"No face found in '{candidate_path}'.")
        continue

    # Use the first face found in the candidate image for comparison
    candidate_face_encoding = candidate_encodings[0]
    
    # Method 1: Euclidean Distance
    euclidean_distance = euclidean(target_face_encoding, candidate_face_encoding)

    # Method 2: Mean Squared Error (MSE)
    mse_score = np.mean((target_face_encoding - candidate_face_encoding)**2)

    # --- Part 4: Display results and provide a conclusion ---
    print(f"  Euclidean Distance: {euclidean_distance:.4f}")
    print(f"  Mean Squared Error (MSE): {mse_score:.6f}")
    
    # These thresholds are a good starting point but can be adjusted.
    if euclidean_distance < 0.5:
        print("  Conclusion: The faces are very likely the same person.")
    elif 0.5 <= euclidean_distance < 0.7:
        print("  Conclusion: The faces are similar.")
    elif 0.7 <= euclidean_distance < 0.9:
        print("  Conclusion: The faces are slightly different.")
    else:
        print("  Conclusion: The faces are very likely of different people.")

print("\n--- End of comparison ---")
print("Note: Lower values for both scores indicate higher similarity.")
```
